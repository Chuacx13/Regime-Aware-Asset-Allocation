```python
from RegimeAwareAllocator import RegimeAwareAllocator
```

```python
allocator = RegimeAwareAllocator(start_date="2005-01-01", out_of_sample_years=2, verbose=False)
allocator.fetch_asset_data()
allocator.preprocess_features()
best_k = allocator.optimize_n_components_by_bic(window=1000)
allocator.identify_regimes(window=1000, n_components=best_k)          
allocator.combine_regimes_to_features() 
allocator.convert_log_to_simple_returns()
allocator.plot_regime_distribution() 
allocator.plot_transition_likelihood() 
allocator.shift_returns()
weights = allocator.compute_all_regime_weights()

allocator.calculate_regime_aware_max_sharpe_is(weights_df=weights)
allocator.calculate_rolling_rpp_is()
allocator.calculate_rolling_max_sharpe_is()
allocator.calculate_6040_is()
print("In-Sample Performance:")
allocator.plot_cumulative_wealth(allocator.is_returns)
allocator.print_metrics(allocator.is_returns)

allocator.calculate_regime_aware_max_sharpe_oos(weights_df=weights)
allocator.calculate_rolling_rpp_oos()
allocator.calculate_rolling_max_sharpe_oos()
allocator.calculate_6040_oos()
print("Out-of-Sample Performance:")
allocator.plot_cumulative_wealth(allocator.oos_returns)
allocator.print_metrics(allocator.oos_returns)
```

```python
sims = allocator.simulate_block_bootstrap_mc(n_sims=5000)
mc_summary = allocator.summarize_mc(sims)
mc_vs_oos = allocator.compare_mc_vs_actual_oos(sims)

allocator.plot_mc_wealth_fan(sims, strategy_name='regime_max_sharpe')
allocator.plot_mc_sharpe_hist(sims, strategy_name='regime_max_sharpe')
```
